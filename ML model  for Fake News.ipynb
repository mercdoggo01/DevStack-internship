#### We have no inconsistent values hence we can proceed

```python
labels = news_df.label
```

#### do this if u don't have sklearn

#### We are going to use sklearn here if you want more information of sklearn.model_selection use this ['https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html']

```python
from sklearn.model_selection import train_test_split
```

#### Training

### The need for training data you may refer 'https://appen.com/blog/training-data/'


```python
# basic concepts of ML taking test size 0.25 and random state 20
x_train,x_test,y_train,y_test = train_test_split(news_df["text"],labels,test_size=0.25,random_state = 20)
```

```python
x_train.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
```

#### Intializing the vectorizer

```python
vc = TfidfVectorizer(stop_words = 'english', max_df = 0.7)
```

#### Fit and Transform

```python
tf_train  = vc.fit_transform(x_train)
tf_test = vc.transform(x_test)
```

#### Intializing passive aggressive classifier

```python
pgc = PassiveAggressiveClassifier(max_iter=50)
pgc.fit(tf_train,y_train)
```

#### Predicition

#### For more info on prediction and passive aggresive classifier use 'https://www.geeksforgeeks.org/passive-aggressive-classifiers/'

```python
from sklearn.metrics import accuracy_score, confusion_matrix
pred = pgc.predict(tf_test)
```

```python
score = accuracy_score(y_test,pred)
```

```python
print("Accuracy is {} %".format(round(score*100,2)))
```

#### confusion matrix

#### A confusion matrix is a table that is often used to describe the performance of a classification model (or "classifier") on a set of test data for which the true values are known.

```python
confusion_matrix(y_test,pred,labels=['FAKE','REAL'])
```
